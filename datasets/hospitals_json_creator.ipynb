```python
import pandas as pd
```

```python
data = pd.read_json('hospitals_norm.json')

print data.columns
```

```python
def centroid(vertexes):
     _x_list = [vertex[0] for vertex in vertexes]
     _y_list = [vertex[1] for vertex in vertexes]
     _len = len(vertexes)
     _x = sum(_x_list) / _len
     _y = sum(_y_list) / _len
     return(_x, _y)
```

```python
output = '[\n'
for (index, row) in data.iterrows():
    output += "{\n" + "\t\"type\":" + "\"hospitals\",\n"
    center = centroid(row['geoData']['coordinates'])
    output += "\t\"lat\": " + str(center[1]) + ",\n"
    output += "\t\"lon\": " + str(center[0])
    output += "\n},\n"
output = output[:-2] + "]"
open('hospitals.json', 'w').write(output)
```
