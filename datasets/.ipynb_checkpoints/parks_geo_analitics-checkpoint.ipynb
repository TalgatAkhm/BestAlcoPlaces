```python
# geo_data = df['geodata']
# obj_type = df['object_type']
# new_columns =[i for i in list(df.columns) if i == 'object_type' and i =='geoData']
df2 = df['geoData']
df2.head()
```

```python
a = np.array(df2.values)
res = list()
for dic in a:
#     print(dic)
    tmp = {'object_type': 'parks'}
    for o in dic.items():
#         print(o)
        if(o[0] == 'type'):
            tmp.update({'polygon_type': o[1]})
        elif(o[0] == 'coordinates'):
#             points = o[1]
            for p in o[1]:
#                 print(p)
                p_arr = []
                for t in p:
#                     print(t)
#                     print()
                    p_arr.append({'lat' : t[1], 'lon': t[0]})
            tmp.update({'coordinates' : p_arr})
    res.append(tmp)
res
```

```python
file = open("result/parks.json", "w")

file.write(json.dumps(res))
```
